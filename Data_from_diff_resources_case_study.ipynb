TASK 5
Explore each dataset.
Find the following:

● Column names
● Data types
● Missing values
● Duplicate records

```python
def explore_dataset(df, name):
    print(f"\n--- {name} ---")
    print("Shape (rows, columns):", df.shape)
    print("Column names:", df.columns.tolist())
    print("Data types:\n", df.dtypes)
    print("Missing values:\n", df.isnull().sum())
    print("Duplicate rows:", df.duplicated().sum())

# Explore all datasets
explore_dataset(gdp, "Excel Dataset")
explore_dataset(population, "CSV Dataset")
explore_dataset(internet_users, "JSON Dataset")
explore_dataset(literacy_rate, "XML Dataset")
```

Part C - DATA TRANSFORMATION

TASK 6

```python
gdp = gdp.rename(columns={'Country':'country'})
literacy_rate = literacy_rate.rename(columns={'name':'country'})
```

```python
set(gdp['country']) == set(internet_users['country']) == set(population['country']) == set(literacy_rate['country'])
```

TASK 7

```python
population['population'] = pd.to_numeric(population['population'])
gdp['GDP'] = pd.to_numeric(gdp['GDP'])
internet_users['internet_users'] = pd.to_numeric(internet_users['internet_users'])
literacy_rate['literacy_rate'] = pd.to_numeric(literacy_rate['literacy_rate'])


print("Population dtype:", population['population'].dtype)
print("GDP dtype:", gdp['GDP'].dtype)
print("Internet Users dtype:", internet_users['internet_users'].dtype)
print("Literacy Rate dtype:", literacy_rate['literacy_rate'].dtype)
```

TASK 8


```python
merged_data = population.merge(gdp, on='country') \
                             .merge(internet_users, on='country') \
                             .merge(literacy_rate, on='country')

merged_data.head()
```

TASK 9

```python
merged_data["Internet_penetration(%)"] = round((merged_data["internet_users"] / merged_data["population"]) * 100,2)
merged_data["GDP_per_capita"] = round(merged_data["GDP"] / merged_data["population"],2)
merged_data.head()

```

PART D - LOAD DATA

TASK 10

```python
import sqlite3
conn = sqlite3.connect('country_data.db')
merged_data.to_sql("country_statistics",conn,if_exists="replace",index=False)
print("Database and 'country_statistics' table succesfully")
```

TASK 11

```python
merged_check = pd.read_sql("select * from country_statistics limit 5", conn)
print(merged_check)
```

PART E -DATA ANALYSIS USING PANDAS

```python
#1  List the Top 5 countries by GDP.
top5_gdp = merged_data.sort_values(by="GDP", ascending=False).head(5)
print(top5_gdp[["country","GDP"]])
```

```python
#2 List the Top 5 countries by population.
top_population = merged_data.sort_values(by="population",ascending=False).head(5)
print(top_population[["country","population"]])
```

```python
#3 Calculate the average literacy rate.
average_literacy = merged_data["literacy_rate"].mean().round(2)
print("Average Literacy Rate:", average_literacy)
```

```python
#4  List all countries having a literacy rate greater than 90%.
high_literacy = merged_data[merged_data["literacy_rate"] > 90]
print(high_literacy[["country", "literacy_rate"]])
```

```python
#5 List all countries having an internet penetration greater than 70%.
high_internet = merged_data[merged_data["Internet_penetration(%)"] > 70]
print(high_internet[["country", "Internet_penetration(%)"]])
```

```python
#6 List the Top 5 countries by GDP per capita.
top5_gdp_per_capita = merged_data.sort_values(by="GDP_per_capita", ascending=False).head(5)
print(top5_gdp_per_capita[["country", "GDP_per_capita"]])
```

```python
#7 Calculate the total population represented in the dataset.
total_population = merged_data["population"].sum()
print("Total Population in Dataset:", total_population)
```

```python
#8 Calculate the average internet penetration across all countries.
average_internet_penetration = merged_data["Internet_penetration(%)"].mean().round(2)
print("Average Internet Penetration (%):", average_internet_penetration)
```

```python
#9 Identify the country having the highest number of internet users.
max_internet_users = merged_data.loc[merged_data["internet_users"].idxmax()]
print("Country with highest internet users:", max_internet_users["country"])
print("Number of internet users:", max_internet_users["internet_users"])
```

```python
#10  Find the correlation between Literacy Rate and Internet Penetration.
correlation = merged_data["literacy_rate"].corr(merged_data["Internet_penetration(%)"])
print("Correlation between Literacy Rate and Internet Penetration:", round(correlation, 2))
```

PART F - DATA ANALYSIS USING SQL

```python
#11 Display the Top 5 countries by GDP.
query = """select country, GDP from country_statistics order by GDP desc limit 5;"""
top5_gdp = pd.read_sql(query, conn)
print(top5_gdp)
```

```python
#12 Find the average literacy rate.
query = """SELECT ROUND(AVG(literacy_rate), 2) AS Average_Literacy_Rate FROM country_statistics;"""
avg_literacy = pd.read_sql(query, conn)
print(avg_literacy)
```

```python
#13 Display all countries having literacy rate above 90%.
query = """SELECT country, literacy_rate FROM country_statistics WHERE literacy_rate > 90 ORDER BY literacy_rate DESC;"""
high_literacy = pd.read_sql(query, conn)
print(high_literacy)
```

```python
#14 Find the Top 5 countries with the highest internet penetration.
top5_internet = merged_data.sort_values(by="Internet_penetration(%)", ascending=False).head(5)
print(top5_internet[["country", "Internet_penetration(%)"]])
```

```python
#15 Calculate the total population.
total_population = merged_data["population"].sum()
print("Total Population in Dataset:", total_population)
```

```python
#16 Find the country with the highest GDP per capita.
highest_gdp_per_capita = merged_data.loc[merged_data["GDP_per_capita"].idxmax()]
print("Country with highest GDP per capita:", highest_gdp_per_capita["country"])
print("GDP per capita:", highest_gdp_per_capita["GDP_per_capita"])
```
